Part 1: Data Loading and Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_header = pd.read_csv('STONYBRK_20240531_HEADER.csv')
df_line = pd.read_csv('STONYBRK_20240531_LINE.csv')
df_code = pd.read_csv('STONYBRK_20240531_CODE.csv')
```

```python
print("DataFrame dimensions:", df_header.shape)

print("\nPreview of the first five rows:")
display(df_header.head())

print("\nColumn types:")
print(df_header.info())

print("\nCount of missing values in each column:")
print(df_header.isna().sum())

print("\nSummary statistics:")
print(df_header.describe(include='all'))
```

```python
print("Dataset size:", df_line.shape)

print("\nSample of the first five entries:")
display(df_line.head())

print("\nData types for each column:")
df_line.info()

print("\nMissing value summary:")
print(df_line.isna().sum())

print("\nStatistical overview:")
print(df_line.describe(include='all'))
```

```python
print("Dimensions of the CODE dataset:", df_code.shape)

print("\nPreview of the first five rows:")
display(df_code.head())

print("\nColumn information:")
df_code.info()

print("\nMissing value check:")
print(df_code.isna().sum())

print("\nDescriptive summary:")
print(df_code.describe(include='all'))
```

```python
# 1. Count how many distinct claims appear in the dataset
claim_total = df_header["ProspectiveClaimId"].nunique()
print("Total unique claims in the dataset:", claim_total)

# 2. Determine the time span covered by the claims
df_header["ServiceFromDate"] = pd.to_datetime(df_header["ServiceFromDate"])
df_header["ServiceToDate"]   = pd.to_datetime(df_header["ServiceToDate"])

start_date = df_header["ServiceFromDate"].min().date()
end_date   = df_header["ServiceToDate"].max().date()

print(f"Claim service dates range from {start_date} through {end_date}")

# 3. Compute average number of service lines per claim
avg_lines = df_line.groupby("ProspectiveClaimId").size().mean()
print("Average number of service lines per claim:", round(avg_lines, 2))

# 4. Compute average number of diagnosis codes per claim
avg_dx = df_code.groupby("ProspectiveClaimId").size().mean()
print("Average number of diagnosis codes per claim:", round(avg_dx, 2))
```

Part 2: Relational Data Analysis


```python
## Question 1: Provider Analysis

# Count how many claims each billing provider submitted
provider_summary = (
    df_header
    .groupby(["BillingProviderNPI", "BillingProvFirstName"])
    .size()
    .reset_index(name="TotalClaims")
)

# Identify the top 5 providers
top_providers = provider_summary.sort_values("TotalClaims", ascending=False).head(5)

# Create a cleaner display name
top_providers["ProviderName"] = top_providers["BillingProvFirstName"]

# Show the results in a readable format
print(top_providers[["ProviderName", "BillingProviderNPI", "TotalClaims"]])

# Visualization
plt.figure(figsize=(8, 6))
sns.barplot(
    data=top_providers,
    x="ProviderName",
    y="TotalClaims",
    linewidth=1,
    edgecolor="black"
)
plt.title("Top 5 Billing Providers by Number of Claims")
plt.xlabel("Provider")
plt.ylabel("Number of Claims")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
## Question 2: Payer Mix Analysis

# Determine the top five payers based on claim frequency
payer_counts = (
    df_header["PrimaryPayerName"]
    .value_counts()
    .iloc[:5]
)

print("Top five primary payers by number of submitted claims:")
print(payer_counts)

# Compute each payer's share of total claims
claim_total = df_header.shape[0]
payer_share = (payer_counts / claim_total * 100).round(2)

# Create a tidy table for reporting
payer_summary = pd.DataFrame({
    "Payer": payer_counts.index,
    "Claims": payer_counts.values,
    "PercentOfTotal": payer_share.values
})

print("\nPayer distribution summary:")
print(payer_summary)

# Visualization
plt.figure(figsize=(8, 6))
sns.barplot(
    data=payer_summary,
    x="Payer",
    y="Claims",
    edgecolor="black"
)
plt.title("Top 5 Primary Payers by Claim Volume")
plt.xlabel("Primary Payer")
plt.ylabel("Number of Claims")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
## Question 3: Common Diagnoses

# Safely compute the top 10 ICD-10 diagnosis codes
top_dx = df_code["CodeValue"].value_counts().reset_index()
top_dx.columns = ["ICD10", "Count"]   # force rename to avoid errors
top_dx = top_dx.head(10)

print("Most common diagnosis codes:")
print(top_dx)

# Visualization - horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_dx,
    y="ICD10",
    x="Count",
    edgecolor="black",
    color="skyblue"
)

plt.title("Top 10 Most Frequently Used Diagnosis Codes", fontsize=14)
plt.xlabel("Number of Occurrences")
plt.ylabel("ICD-10 Code")
plt.tight_layout()
plt.show()
```

```python
## Question 4: Common Procedures

# Get frequency of HCPCS procedure codes
procedure_counts = df_line["HCPCS"].value_counts().reset_index()

# Force column names to avoid rename errors
procedure_counts.columns = ["HCPCS_Code", "Count"]

# Take the top 10
procedure_counts = procedure_counts.head(10)

print("Top 10 procedure codes:")
print(procedure_counts)

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(
    data=procedure_counts,
    x="HCPCS_Code",
    y="Count",
    edgecolor="black",
    color="cornflowerblue"
)

plt.title("Most Frequently Billed Procedure Codes", fontsize=14)
plt.xlabel("Procedure (HCPCS)")
plt.ylabel("Frequency")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

```

```python
## Question 5: Service Location Analysis

# Count claims by place of service
pos_counts = df_header["PlaceOfService"].value_counts().reset_index()
pos_counts.columns = ["PlaceOfService", "Claims"]

print("Claim counts by Place of Service:")
print(pos_counts)

# Total claims in the dataset
total_claims = df_header.shape[0]

# Add percentage column
pos_counts["Percent"] = (pos_counts["Claims"] / total_claims * 100).round(2)

print("\nPlace of Service with Percentages:")
print(pos_counts)

# OPTIONAL reference codes (these depend on your dataset)
print("\nCommon POS Codes:")
print("21 = Inpatient Hospital")
print("11 = Office")
print("22 = Outpatient Hospital (On Campus)")
print("23 = Emergency Room")

# Check if Facility_Location column exists before using it
if "Facility_Location" in df_header.columns:
    facility_summary = (
        df_header["Facility_Location"]
        .value_counts()
        .reset_index()
    )
    facility_summary.columns = ["Facility_Location", "Claims"]
    facility_summary["Percent"] = (
        facility_summary["Claims"] / total_claims * 100
    ).round(2)

    print("\nClaims by Facility Location:")
    print(facility_summary)

    # Filter for specific facility types ONLY IF they exist
    target_locations = ["INPATIENT", "DOCTOR'S OFFICE"]
    filtered = facility_summary[
        facility_summary["Facility_Location"].isin(target_locations)
    ]

    print("\nFiltered Facility Locations (Inpatient & Office):")
    print(filtered)
else:
    print("\n⚠️ 'Facility_Location' column not found in df_header — skipping facility analysis.")
```

Part 3: Advanced Analysis with Joins


```python
## Question 6: Claims with High Service Line Counts

# Join claim header and line-level detail
claims_with_lines = df_header.merge(
    df_line,
    on="ProspectiveClaimId",
    how="inner"
)

# Count how many service lines each claim has
line_counts = (
    claims_with_lines
    .groupby("ProspectiveClaimId")
    .size()
    .reset_index(name="NumLines")
)

print("Service line count per claim:")
print(line_counts)

# Keep only claims that have 5 or more lines
busy_claims = line_counts[line_counts["NumLines"] >= 5]
print("\nNumber of claims with 5+ service lines:", len(busy_claims))

# Compute total charges per claim
charges_by_claim = (
    claims_with_lines
    .groupby("ProspectiveClaimId")["Charges"]
    .sum()
    .reset_index(name="TotalCharges")
)

# Pull provider info from the header table (drop duplicates just in case)
if "BillingProvFirstName" in df_header.columns:
    provider_info = (
        df_header[["ProspectiveClaimId", "BillingProvFirstName"]]
        .drop_duplicates()
        .rename(columns={"BillingProvFirstName": "ProviderName"})
    )
else:
    provider_info = df_header[["ProspectiveClaimId"]].drop_duplicates()
    provider_info["ProviderName"] = "Unknown"

# Combine line counts, total charges, and provider name
high_line_summary = (
    busy_claims
    .merge(charges_by_claim, on="ProspectiveClaimId", how="left")
    .merge(provider_info, on="ProspectiveClaimId", how="left")
)

print("\nClaims with 5 or more service lines:")
print(high_line_summary[["ProspectiveClaimId", "ProviderName", "NumLines", "TotalCharges"]])
```

```python
## Question 7: Diagnosis–Procedure Combinations

# Step 1: Combine HEADER + LINE + CODE into a single table
claims_full = (
    df_header
    .merge(df_line, on="ProspectiveClaimId", how="inner")
    .merge(df_code, on="ProspectiveClaimId", how="inner")
)

# Step 2: Focus only on claims that include CPT/HCPCS 99291
critical_care_claims = claims_full[claims_full["HCPCS"] == "99291"]

# Safety check: make sure we actually have any 99291 rows
if critical_care_claims.empty:
    print("No claims found with HCPCS code 99291.")
else:
    # Step 3: Count which diagnosis codes appear most often with 99291
    dx_for_99291 = (
        critical_care_claims["CodeValue"]
        .value_counts()
        .reset_index()
    )
    dx_for_99291.columns = ["DiagnosisCode", "Count"]

    print("Diagnosis codes most commonly associated with CPT 99291:")
    print(dx_for_99291.head(10))

    # Step 4: Highlight the single most frequent diagnosis code
    top_row = dx_for_99291.iloc[0]
    print(f"\nMost common diagnosis code with 99291: {top_row['DiagnosisCode']} (Count: {top_row['Count']})")
```

```python
## Question 8: Charges by Payer

# Step 1: Combine claim header and line-level charges
claims_payer = df_header.merge(
    df_line,
    on="ProspectiveClaimId",
    how="inner"
)

# Step 2: Roll up to the claim level (one row per claim)
claim_level = (
    claims_payer
    .groupby("ProspectiveClaimId")
    .agg(
        TotalCharges=("Charges", "sum"),
        Payer=("PrimaryPayerName", "first")
    )
    .reset_index()
)

print("Sample of claim-level charges with payer:")
print(claim_level.head())

# Step 3: Aggregate by payer
payer_summary = (
    claim_level
    .groupby("Payer")
    .agg(
        TotalCharges=("TotalCharges", "sum"),
        NumClaims=("ProspectiveClaimId", "count")
    )
    .reset_index()
)

# Step 4: Compute average charge per claim for each payer
payer_summary["AvgChargePerClaim"] = (
    payer_summary["TotalCharges"] / payer_summary["NumClaims"]
).round(2)

# Step 5: Sort by total charges in descending order
payer_summary = payer_summary.sort_values("TotalCharges", ascending=False)

# Step 6: Display top 10 payers
print("\nTop 10 payers by total billed charges:")
print(payer_summary.head(10))
```

Part 4: Creative Analysis


```python
## Custom Analysis: Common Diagnosis Code Pairs

# Step 1: Create all diagnosis-code pairs within each claim (self-merge on claim ID)
dx_pairs = df_code.merge(
    df_code,
    on="ProspectiveClaimId",
    suffixes=("_A", "_B")
)

# Step 2: Remove self-pairs where the diagnosis code is identical
dx_pairs = dx_pairs[dx_pairs["CodeValue_A"] != dx_pairs["CodeValue_B"]]

# Step 3: Canonicalize the pair ordering so (A, B) and (B, A) are treated as the same pair
dx_pairs["CodeLow"]  = dx_pairs[["CodeValue_A", "CodeValue_B"]].min(axis=1)
dx_pairs["CodeHigh"] = dx_pairs[["CodeValue_A", "CodeValue_B"]].max(axis=1)

# Step 4: Count how often each unique pair of codes appears together on a claim
pair_stats = (
    dx_pairs
    .groupby(["CodeLow", "CodeHigh"])
    .size()
    .reset_index(name="Frequency")
    .sort_values("Frequency", ascending=False)
)

print("Top 10 most common diagnosis code pairs:")
print(pair_stats.head(10))
```
